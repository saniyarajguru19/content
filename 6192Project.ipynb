# UNIVARIATE ANALYSIS

```python
plt.figure(figsize = (10, 10))
plt.style.use('seaborn-white')
ax=plt.subplot(221)
plt.boxplot(df['math score'])
ax.set_title('Math score')
ax=plt.subplot(222)
plt.boxplot(df['reading score'])
ax.set_title('Reading Score')
ax=plt.subplot(223)
plt.boxplot(df['writing score'])
ax.set_title('Writing Score')
```

Conclusion:The provided code creates a single figure with three side-by-side box plots, each representing the distribution of scores in different subjects: "Math Score," "Reading Score," and "Writing Score."

# BIVARIATE ANALYSIS

```python
score = df[['math score', 'reading score', 'writing score']]
score.head()
```

Conclusion:The code score = df[['math score', 'reading score', 'writing score']] creates a new DataFrame named score by selecting and extracting three specific columns from the original DataFrame df. These columns are "math score," "reading score," and "writing score." The head() method is then used to display the first five rows of this new DataFrame.

```python
# Scatter Plot between Reading Score and Writing score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='reading score', y='writing score',hue="parental level of education", data=df, color='skyblue')
plt.title("Scatter Plot: Reading score vs. Writing score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

Conclusion:The provided code creates a scatter plot using Seaborn to visualize the relationship between "reading score" and "writing score" for different levels of "parental level of education" from a pandas DataFrame (df).

CORRELATION BETWEEN SUBJECT SCORES

```python
corr = score.corr()
sns.heatmap(corr, annot = True, fmt=".2f", cmap='coolwarm')
```

Conclusion:The provided code computes the correlation matrix between the numeric columns (math score, reading score, and writing score) in the DataFrame score, and then it creates a heatmap visualization of the correlation matrix using Seaborn (sns).

```python
df['mean score'] = score.mean(axis=1)
df = df.drop(['math score', 'reading score', 'writing score'], axis=1)
df.head()
```

Conclusion:The provided code adds a new column named "mean score" to the DataFrame df. This new column contains the mean (average) score computed from the "math score," "reading score," and "writing score" columns. Afterward, it drops these three columns from the DataFrame. Finally, it displays the first five rows of the modified DataFrame using df.head()

```python
ax = sns.barplot(data=df, y='parental level of education', x='mean score', orient="h")
plt.title("Mean score by parental level of education")
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_width():.2f}', (p.get_x() + (2+p.get_width()), p.get_y() + p.get_height()/ 2.), ha='left', va='center', fontsize=11, color='black', xytext=(5, 0), textcoords='offset points')
```

Conclusion:The provided code is creating a horizontal bar plot using Seaborn (sns.barplot) to visualize the mean scores grouped by the "parental level of education" from a pandas DataFrame (df). It also includes annotations on the bars to display the mean scores.
